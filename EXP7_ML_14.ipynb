## Implement the cost function

### Subtask:
Define a cost function (e.g., Mean Squared Error) to measure the difference between predicted and actual values.


**Reasoning**:
Define the mean squared error cost function as requested in the instructions.



```python
import numpy as np

def mean_squared_error(X, y, weights, bias):
    """
    Calculates the Mean Squared Error (MSE) cost.

    Args:
        X: Feature matrix (numpy array).
        y: Target vector (numpy array).
        weights: Weight vector (numpy array).
        bias: Bias term (float).

    Returns:
        The Mean Squared Error (float).
    """
    predictions = np.dot(X, weights) + bias
    mse = np.mean((predictions - y) ** 2)
    return mse

# Example usage (optional, for testing the function)
# You can call this function later with your X, y, weights, and bias
# cost = mean_squared_error(X, y, weights, bias)
# print(f"Initial cost: {cost}")
```

## Implement the gradient descent algorithm

### Subtask:
Write the code to iteratively update the model parameters to minimize the cost function.


**Reasoning**:
Define the gradient descent function to iteratively update the model parameters.



```python
def gradient_descent(X, y, weights, bias, learning_rate, iterations):
    """
    Performs gradient descent to minimize the cost function.

    Args:
        X: Feature matrix (numpy array).
        y: Target vector (numpy array).
        weights: Initial weight vector (numpy array).
        bias: Initial bias term (float).
        learning_rate: The learning rate for gradient descent (float).
        iterations: The number of iterations to perform gradient descent (int).

    Returns:
        A tuple containing the final weights, final bias, and a list of costs
        at each iteration.
    """
    m = len(y)
    cost_history = []

    for i in range(iterations):
        # Calculate predictions
        predictions = np.dot(X, weights) + bias

        # Calculate gradients
        weights_gradient = (1/m) * np.dot(X.T, (predictions - y))
        bias_gradient = (1/m) * np.sum(predictions - y)

        # Update weights and bias
        weights = weights - learning_rate * weights_gradient
        bias = bias - learning_rate * bias_gradient

        # Calculate and store cost (optional)
        cost = mean_squared_error(X, y, weights, bias)
        cost_history.append(cost)

    return weights, bias, cost_history
```
